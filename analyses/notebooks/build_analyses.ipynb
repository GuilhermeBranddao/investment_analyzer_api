- Se o preço está acima das médias móveis → tendência de alta
- Se está abaixo → tendência de baixa
- Se está cruzando frequentemente → tendência lateral


````python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Baixar dados históricos da ITUB4.SA
ticker = "ITUB4.SA"
df = yf.download(ticker, start="2025-01-05", end="2025-10-27", period="max")

# Calcular médias móveis para ajudar na identificação de tendência
df["MM20"] = df["Close"].rolling(window=20).mean()
df["MM50"] = df["Close"].rolling(window=50).mean()

# Plotar gráfico com médias móveis
plt.figure(figsize=(24, 7))
plt.plot(df["Close"], label="Preço de Fechamento", color="black")
plt.plot(df["MM20"], label="Média Móvel 20", color="blue", linestyle="--")
plt.plot(df["MM50"], label="Média Móvel 50", color="red", linestyle="--")
plt.title("ITUB4.SA - Tendência com Médias Móveis")
plt.xlabel("Data")
plt.ylabel("Preço (R$)")
plt.legend()
plt.grid(True)
plt.show()
````
